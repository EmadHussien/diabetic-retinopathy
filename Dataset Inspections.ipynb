# Image sizes

```python
from PIL import Image
from os import listdir


def ls(path):
	'''List all visible files'''
	return [f for f in listdir(path) if not f.startswith('.')]

imageSizes = []

for className in ls('./data/subset'):
    classPath = './data/subset/{}/'.format(className)
    for image in ls(classPath):
        with Image.open(classPath + image) as img:
            width, height = img.size
            imageSizes.append((height, width, image))
```

```python
print(imageSizes)
```

```python
import matplotlib.pyplot as plt
widths = [s[1] for s in imageSizes]
print(min(widths), max(widths))
plt.plot(widths)
# plt.xlabel('width')
# plt.ylabel('height')
plt.show()
```

```python
print(sorted(imageSizes))
```

```python
plt.plot(sorted(imageSizes))
plt.xlabel('width')
plt.ylabel('height')
plt.show()
```
